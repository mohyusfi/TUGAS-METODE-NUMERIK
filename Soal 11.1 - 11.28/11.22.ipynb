Persamaan diubah menjadi bentuk matriks `A x = b`, lalu dihitung solusi, transpose, dan invers koefisien.

```python
import numpy as np
np.set_printoptions(precision=6, suppress=True)

A = np.array([[0, -7, 5], [0, 4, 7], [-4, 3, -7]], float)
b = np.array([50, -30, 40], float)
x = np.linalg.solve(A, b)
print("A=\n", A)
print("b=", b)
print("x=", x)
print("A.T=\n", A.T)
print("A^-1=\n", np.linalg.inv(A))
print("cek A@x=", A @ x)
```

Bentuk matriks dipilih setelah semua variabel diletakkan di kiri dan konstanta di kanan.
