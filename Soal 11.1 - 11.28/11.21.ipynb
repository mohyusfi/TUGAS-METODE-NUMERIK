Perintah satu baris MATLAB untuk membentuk matriks augmentasi `[A | I]` adalah:

```matlab
Aug = [A eye(size(A))]
```

Di bawah ini ditunjukkan padanan NumPy.

````python
import numpy as np
np.set_printoptions(precision=6, suppress=True)

A = np.array([[1, 2], [3, 4]], float)
Aug = np.hstack((A, np.eye(A.shape[0])))
print(Aug)
````

`eye(size(A))` membuat matriks identitas berukuran sama dengan `A`, lalu operator kurung siku MATLAB menggabungkannya secara horizontal.
