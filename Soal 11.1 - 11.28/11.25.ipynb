Program Cholesky berdasarkan algoritma pada Fig. 11.3 dan diuji dengan Example 11.2.

```python
import numpy as np
np.set_printoptions(precision=6, suppress=True)

def cholesky_lower(A):
    A = np.array(A, dtype=float)
    n = A.shape[0]
    L = np.zeros_like(A)
    for k in range(n):
        for i in range(k):
            L[k, i] = (A[k, i] - np.dot(L[k, :i], L[i, :i])) / L[i, i]
        L[k, k] = np.sqrt(A[k, k] - np.dot(L[k, :k], L[k, :k]))
    return L

def forward_substitution(L, b):
    b = np.array(b, dtype=float)
    y = np.zeros_like(b)
    for i in range(len(b)):
        y[i] = (b[i] - np.dot(L[i, :i], y[:i])) / L[i, i]
    return y

def back_substitution(U, y):
    y = np.array(y, dtype=float)
    x = np.zeros_like(y)
    for i in range(len(y) - 1, -1, -1):
        x[i] = (y[i] - np.dot(U[i, i + 1:], x[i + 1:])) / U[i, i]
    return x

def solve_cholesky(A, b):
    L = cholesky_lower(A)
    y = forward_substitution(L, b)
    x = back_substitution(L.T, y)
    return L, x

A = np.array([[6, 15, 55], [15, 55, 225], [55, 225, 979]], float)
b = A.sum(axis=1)
L, x = solve_cholesky(A, b)
print("L=\n", L)
print("x untuk b jumlah baris =", x)
print("cek L@L.T=\n", L @ L.T)
```

Karena `b` dibuat dari jumlah baris `A`, solusi ideal adalah `[1, 1, 1]`; hasil program memverifikasi implementasi.
