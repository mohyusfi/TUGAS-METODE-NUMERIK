Tiga set diuji dengan kriteria radius spektral matriks iterasi Gauss-Seidel. Jika `rho < 1`, iterasi konvergen; jika `rho > 1`, iterasi divergen untuk sebagian besar tebakan awal.

```python
import numpy as np
np.set_printoptions(precision=6, suppress=True)

def gauss_seidel(A, b, x0=None, es=5.0, max_iter=100, lam=1.0):
    A = np.array(A, dtype=float)
    b = np.array(b, dtype=float)
    n = len(b)
    x = np.zeros(n) if x0 is None else np.array(x0, dtype=float)
    rows = []
    for it in range(1, max_iter + 1):
        old = x.copy()
        for i in range(n):
            raw = (b[i] - np.dot(A[i, :i], x[:i]) - np.dot(A[i, i + 1:], old[i + 1:])) / A[i, i]
            x[i] = lam * raw + (1 - lam) * old[i]
        ea = np.full(n, np.nan)
        mask = x != 0
        ea[mask] = np.abs((x[mask] - old[mask]) / x[mask]) * 100
        rows.append([it, *x, np.nanmax(ea)])
        if np.nanmax(ea) < es:
            break
    return x, np.array(rows)

def jacobi(A, b, x0=None, es=5.0, max_iter=100):
    A = np.array(A, dtype=float)
    b = np.array(b, dtype=float)
    n = len(b)
    x = np.zeros(n) if x0 is None else np.array(x0, dtype=float)
    rows = []
    for it in range(1, max_iter + 1):
        old = x.copy()
        for i in range(n):
            x[i] = (b[i] - np.dot(A[i, :i], old[:i]) - np.dot(A[i, i + 1:], old[i + 1:])) / A[i, i]
        ea = np.full(n, np.nan)
        mask = x != 0
        ea[mask] = np.abs((x[mask] - old[mask]) / x[mask]) * 100
        rows.append([it, *x, np.nanmax(ea)])
        if np.nanmax(ea) < es:
            break
    return x, np.array(rows)

sets = {
    "Set One": (np.array([[8, 3, 1], [-6, 0, 7], [2, 4, -1]], float), np.array([12, 1, 5], float), [0, 2, 1]),
    "Set Two": (np.array([[1, 1, 5], [1, 4, -1], [3, 1, -1]], float), np.array([7, 4, 4], float), [2, 1, 0]),
    "Set Three": (np.array([[2, 3, 5], [-2, 4, -5], [0, 2, -1]], float), np.array([7, -3, 1], float), [1, 2, 0]),
}
for name, (A0, b0, order) in sets.items():
    A = A0[order]
    b = b0[order]
    D = np.diag(np.diag(A))
    L = np.tril(A, -1)
    U = np.triu(A, 1)
    T = -np.linalg.inv(D + L) @ U
    rho = max(abs(np.linalg.eigvals(T)))
    x, hist = gauss_seidel(A, b, es=5, max_iter=50)
    print("\n", name)
    print("urutan baris =", order)
    print("rho =", rho)
    print("x pendekatan =", x)
    print("iterasi =", len(hist))
```

Dengan penyusunan ulang baris yang tepat, ketiga set memiliki `rho < 1`, sehingga dapat diselesaikan oleh Gauss-Seidel. Jika memakai urutan awal tanpa penyusunan ulang, Set One bermasalah karena ada diagonal nol, dan Set Two divergen karena `rho > 1`.
