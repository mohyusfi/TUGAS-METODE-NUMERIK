Sistem pentadiagonal diselesaikan efisien tanpa pivoting dengan eliminasi Gaussian yang hanya menyentuh pita lima diagonal. Kasus uji mengikuti matriks pada soal.

```python
import numpy as np
np.set_printoptions(precision=6, suppress=True)

def solve_pentadiagonal(A, b):
    A = np.array(A, dtype=float).copy()
    b = np.array(b, dtype=float).copy()
    n = len(b)
    for k in range(n - 1):
        last_row = min(n, k + 3)
        for i in range(k + 1, last_row):
            if A[i, k] == 0:
                continue
            factor = A[i, k] / A[k, k]
            last_col = min(n, k + 3)
            A[i, k:last_col] -= factor * A[k, k:last_col]
            b[i] -= factor * b[k]
    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        last_col = min(n, i + 3)
        x[i] = (b[i] - np.dot(A[i, i + 1:last_col], x[i + 1:last_col])) / A[i, i]
    return x

A = np.array([
    [8, -2, -1, 0, 0],
    [-2, 9, -4, -1, 0],
    [-1, -3, 7, -1, -2],
    [0, -4, -2, 12, -5],
    [0, 0, -7, -3, 15],
], float)
b = np.array([5, 2, 1, 1, 5], float)
x = solve_pentadiagonal(A, b)
print("x =", x)
print("cek np.linalg.solve =", np.linalg.solve(A, b))
print("residual =", A @ x - b)
```

Karena elemen nonnol hanya berada pada dua diagonal di bawah dan dua diagonal di atas diagonal utama, eliminasi dibatasi pada pita tersebut sehingga lebih efisien daripada matriks penuh.
