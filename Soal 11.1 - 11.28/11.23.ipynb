Jumlah operasi Thomas dibandingkan dengan eliminasi Gauss naif. Dari algoritma Thomas, total operasi aritmetika kira-kira `8n - 7`; untuk eliminasi Gauss naif orde dominannya `n^3/3`.

```python
import numpy as np
np.set_printoptions(precision=6, suppress=True)

n = np.arange(2, 21)
thomas_ops = 8*n - 7
gauss_ops = n**3 / 3
print(" n | Thomas | Gauss naif")
for i, t, g in zip(n, thomas_ops, gauss_ops):
    print(f"{i:2d} | {t:7.0f} | {g:10.1f}")

# Data ini dapat diplot dengan matplotlib jika tersedia:
# import matplotlib.pyplot as plt
# plt.plot(n, thomas_ops, 'o-', label='Thomas')
# plt.plot(n, gauss_ops, 's-', label='Gauss naif')
# plt.yscale('log'); plt.xlabel('n'); plt.ylabel('operasi'); plt.legend(); plt.grid(True)
```

Pertumbuhan Thomas linear terhadap `n`, sedangkan eliminasi Gauss tumbuh kubik. Untuk sistem tridiagonal besar, perbedaan biayanya sangat besar.
