Persamaan diferensial keadaan tunak

`0 = D c'' - U c' - k c`

diubah menjadi sistem aljabar dengan beda hingga terpusat pada `dx = 2`, `x = 0..10`, dan kondisi batas `c(0)=80`, `c(10)=20`.

```python
import numpy as np
np.set_printoptions(precision=6, suppress=True)

D, U, k, dx = 2.0, 1.0, 0.2, 2.0
x_nodes = np.arange(0, 10 + dx, dx)
left_bc, right_bc = 80.0, 20.0
n = len(x_nodes) - 2
lower = D/dx**2 + U/(2*dx)
diag = -2*D/dx**2 - k
upper = D/dx**2 - U/(2*dx)
A = np.zeros((n, n))
b = np.zeros(n)
for i in range(n):
    A[i, i] = diag
    if i > 0:
        A[i, i-1] = lower
    else:
        b[i] -= lower * left_bc
    if i < n - 1:
        A[i, i+1] = upper
    else:
        b[i] -= upper * right_bc
c_inner = np.linalg.solve(A, b)
c = np.r_[left_bc, c_inner, right_bc]
print("A=\n", A)
print("b=", b)
print("x =", x_nodes)
print("c =", c)
print("data plot (x,c)=")
print(np.column_stack((x_nodes, c)))
```

Koefisien beda hingga menghasilkan sistem tridiagonal untuk empat titik interior. Data `(x, c)` pada output dapat langsung diplot sebagai konsentrasi terhadap jarak.
