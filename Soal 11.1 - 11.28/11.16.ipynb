Hitung solusi, invers, dan condition number berbasis row-sum norm (`inf` norm) tanpa scaling.

```python
import numpy as np
np.set_printoptions(precision=6, suppress=True)

matrices = {
    "a": (np.array([[1, 4, 9], [4, 9, 16], [9, 16, 25]], float), np.array([14, 29, 50], float)),
    "b": (np.array([[1, 4, 9, 16], [4, 9, 16, 25], [9, 16, 25, 36], [16, 25, 36, 49]], float), np.array([30, 54, 86, 126], float)),
}
for label, (A, b) in matrices.items():
    n = A.shape[0]
    rank = np.linalg.matrix_rank(A)
    cond_inf = np.linalg.cond(A, p=np.inf)
    x_expected = np.ones(n)
    print("\nBagian", label)
    print("rank =", rank, "dari", n)
    print("A @ ones =", A @ x_expected)
    print("condition number row-sum =", cond_inf)

    if rank == n:
        invA = np.linalg.inv(A)
        x = np.linalg.solve(A, b)
        print("x =", x)
        print("A^-1 =\n", invA)
    else:
        print("Matriks singular/defisien rank, sehingga invers sejati tidak ada.")
        print("Salah satu solusi yang memenuhi soal adalah x =", x_expected)
        print("residual untuk x=ones =", A @ x_expected - b)
        print("Pseudoinverse Moore-Penrose =\n", np.linalg.pinv(A))
```

Solusi semua `x` seharusnya satu karena ruas kanan adalah jumlah tiap baris. Bagian (b) memberi pelajaran penting: matriks 4x4 sangat ill-conditioned dan terdeteksi defisien rank, sehingga invers sejati tidak stabil/tidak ada secara numerik. Karena itu `x = ones` diverifikasi langsung melalui `A @ ones = b`.
