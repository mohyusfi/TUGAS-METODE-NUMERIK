Model produksi transistor, resistor, dan chip komputer berdasarkan batas bahan baku tembaga, seng, dan kaca.

```python
import numpy as np
np.set_printoptions(precision=6, suppress=True)

A = np.array([[4, 3, 2], [1, 3, 1], [2, 1, 3]], float)
b = np.array([960, 510, 610], float)
x = np.linalg.solve(A, b)
print("[transistor, resistor, chip] =", x)
print("cek kebutuhan bahan =", A @ x)
```

Sistem linear menyatakan konsumsi total bahan. Solusi memberi jumlah komponen yang dapat dibuat minggu tersebut.
