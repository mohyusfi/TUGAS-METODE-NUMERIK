Program ramah-pakai untuk menyelesaikan sistem tridiagonal dengan algoritma Thomas. Program diuji pada Example 11.1.

```python
import numpy as np
np.set_printoptions(precision=6, suppress=True)

def thomas(a, d, c, b):
    a = np.array(a, dtype=float)
    d = np.array(d, dtype=float)
    c = np.array(c, dtype=float)
    b = np.array(b, dtype=float)
    n = len(d)
    l = np.zeros(n - 1)
    u = d.copy()
    r = b.copy()
    for i in range(1, n):
        l[i - 1] = a[i - 1] / u[i - 1]
        u[i] = u[i] - l[i - 1] * c[i - 1]
        r[i] = r[i] - l[i - 1] * r[i - 1]
    x = np.zeros(n)
    x[-1] = r[-1] / u[-1]
    for i in range(n - 2, -1, -1):
        x[i] = (r[i] - c[i] * x[i + 1]) / u[i]
    L = np.eye(n)
    U = np.zeros((n, n))
    for i in range(n):
        U[i, i] = u[i]
        if i < n - 1:
            L[i + 1, i] = l[i]
            U[i, i + 1] = c[i]
    return L, U, r, x

def solve_tridiagonal(lower, diag, upper, rhs):
    _, _, _, x = thomas(lower, diag, upper, rhs)
    return x

x = solve_tridiagonal([-1, -1, -1], [2.04]*4, [-1, -1, -1], [40.8, 0.8, 0.8, 200.8])
print("solusi Example 11.1 =", x)
```

Program menerima tiga diagonal dan ruas kanan, sehingga tidak perlu menyimpan elemen nol pada matriks penuh.
