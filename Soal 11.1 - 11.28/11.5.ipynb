Dekomposisi Cholesky dan solusi sistem simetris untuk koefisien `a0, a1, a2`.

```python
import numpy as np
np.set_printoptions(precision=6, suppress=True)

def cholesky_lower(A):
    A = np.array(A, dtype=float)
    n = A.shape[0]
    L = np.zeros_like(A)
    for k in range(n):
        for i in range(k):
            L[k, i] = (A[k, i] - np.dot(L[k, :i], L[i, :i])) / L[i, i]
        L[k, k] = np.sqrt(A[k, k] - np.dot(L[k, :k], L[k, :k]))
    return L

def forward_substitution(L, b):
    b = np.array(b, dtype=float)
    y = np.zeros_like(b)
    for i in range(len(b)):
        y[i] = (b[i] - np.dot(L[i, :i], y[:i])) / L[i, i]
    return y

def back_substitution(U, y):
    y = np.array(y, dtype=float)
    x = np.zeros_like(y)
    for i in range(len(y) - 1, -1, -1):
        x[i] = (y[i] - np.dot(U[i, i + 1:], x[i + 1:])) / U[i, i]
    return x

A = np.array([[6, 15, 55], [15, 55, 225], [55, 225, 979]], float)
b = np.array([152.6, 585.6, 2488.8], float)
L = cholesky_lower(A)
y = forward_substitution(L, b)
x = back_substitution(L.T, y)
print("L=\n", L)
print("y dari L y=b =", y)
print("[a0, a1, a2] =", x)
print("Cek np.linalg.solve =", np.linalg.solve(A, b))
```

Solusi diperoleh dalam dua tahap substitusi: maju pada `L y=b` dan mundur pada `L.T a=y`.
