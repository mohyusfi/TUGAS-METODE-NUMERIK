Dua persamaan nonlinear diselesaikan dengan Newton-Raphson dua variabel. Lalu tebakan awal integer `x,y = -6..6` dipetakan ke solusi yang dicapai. Untuk persamaan sebagaimana tertulis di PDF, ditemukan empat pasangan solusi nyata.


```python
import numpy as np
np.set_printoptions(precision=6, suppress=True)

def F(v):
    x, y = v
    return np.array([4 - y - 2*x*x, 8 - y*y - 4*x], float)

def J(v):
    x, y = v
    return np.array([[-4*x, -1], [-4, -2*y]], float)

def newton(v0, max_iter=50, tol=1e-10):
    v = np.array(v0, dtype=float)
    for _ in range(max_iter):
        try:
            dv = np.linalg.solve(J(v), -F(v))
        except np.linalg.LinAlgError:
            return None
        v = v + dv
        if np.linalg.norm(dv, ord=np.inf) < tol:
            return v
    return None

seeds = [(2, -4), (1, 2), (-2, -4), (-1, 2)]
sols = []
for s in seeds:
    z = newton(s)
    if z is not None and not any(np.linalg.norm(z - q) < 1e-7 for q in sols):
        sols.append(z)
print("solusi unik =")
for s in sols:
    print(s, "F=", F(s))

counts = {i: 0 for i in range(len(sols))}
failed = 0
for x0 in range(-6, 7):
    for y0 in range(-6, 7):
        z = newton((x0, y0))
        if z is None:
            failed += 1
            continue
        distances = [np.linalg.norm(z - s) for s in sols]
        j = int(np.argmin(distances))
        if distances[j] < 1e-6:
            counts[j] += 1
        else:
            failed += 1
print("jumlah tebakan awal yang menuju tiap solusi =", counts)
print("gagal/di luar klasifikasi =", failed)
```

Untuk persamaan sebagaimana tertulis, terdapat empat pasangan solusi nyata. Pemetaan tebakan awal menunjukkan basin konvergensi Newton; beberapa soal/edisi mungkin menyebut dua pasangan, tetapi bentuk persamaan pada PDF lokal menghasilkan empat perpotongan nyata.
