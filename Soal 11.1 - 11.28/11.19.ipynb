Hilbert 10 dimensi diuji condition number spektralnya. Ruas kanan dibuat sebagai jumlah baris sehingga solusi eksak ideal adalah vektor satu.

```python
import numpy as np
np.set_printoptions(precision=6, suppress=True)

n = 10
H = np.array([[1/(i + j + 1) for j in range(n)] for i in range(n)], float)
b = H.sum(axis=1)
x = np.linalg.solve(H, b)
cond2 = np.linalg.cond(H, p=2)
lost_digits = np.log10(cond2)
print("cond2 =", cond2)
print("perkiraan digit hilang =", lost_digits)
print("x =", x)
print("error absolut maksimum =", np.max(np.abs(x - 1)))
```

Hilbert matrix sangat ill-conditioned. Banyak digit presisi hilang sehingga solusi numerik menyimpang dari vektor satu meskipun ruas kanan dibuat konsisten.
