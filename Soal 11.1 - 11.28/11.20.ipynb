Ulangi analisis condition number untuk matriks Vandermonde 6 dimensi dengan `x = [4, 2, 7, 10, 3, 5]`.

```python
import numpy as np
np.set_printoptions(precision=6, suppress=True)

points = np.array([4, 2, 7, 10, 3, 5], float)
V = np.vander(points, N=6, increasing=True)
b = V.sum(axis=1)
x = np.linalg.solve(V, b)
cond2 = np.linalg.cond(V, p=2)
lost_digits = np.log10(cond2)
print("V=\n", V)
print("cond2 =", cond2)
print("perkiraan digit hilang =", lost_digits)
print("x =", x)
print("error absolut maksimum =", np.max(np.abs(x - 1)))
```

Matriks Vandermonde juga cenderung ill-conditioned, terutama saat nilai titik tersebar. Ruas kanan jumlah baris membuat solusi eksak ideal berupa semua satu.
